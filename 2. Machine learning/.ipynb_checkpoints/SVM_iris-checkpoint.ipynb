```python
from sklearn import svm, datasets
import numpy as np
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
iris = datasets.load_iris()
x = iris.data[:, 0:2]
y = iris.target
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y)
```

```python
# SVC implements one vs one
clf = svm.SVC()
clf.fit(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
def makegrid(x1, x2, h = 0.02):
    x1_min, x1_max = x1.min() - 1, x1.max() + 1
    x2_min, x2_max = x2.min() - 1, x2.max() + 1
    a = np.arange(x1_min, x1_max, h)
    b = np.arange(x2_min, x2_max, h)
    xx, yy = np.meshgrid(a, b) # it will replicate all the points, it will provide us all the points on the grid
    return xx, yy
```

```python
# x1 = np.array([1, 3])
# x2 = np.array([5, 6])
# makegrid(x1, x2)
```

```python
xx, yy = makegrid(x[:, 0], x[:, 1])
predictions = clf.predict(np.c_[xx.ravel(), yy.ravel()]) # ravel will convert our 2 D grid into huge one dimensional array. 
# c_ is used to combine these two pairs. 
plt.scatter(xx.ravel(), yy.ravel(), c = predictions)
plt.show()
```
