```python
def gd(points, learning_rate, num_iterations):
    num_features = points.shape[1] - 1  # Exclude the target column
    m = np.zeros(num_features)  # Vector for slopes
    c = 0  # Scalar for intercept
    for i in range(num_iterations):
        m ,c = step_gradient(points, learning_rate, m, c)
        print(i , " cost:" , cost(points, m, c))
    return m, c
```

```python
def step_gradient(points, learning_rate, m, c):
    num_features = points.shape[1] - 1
    m_slope = np.zeros(num_features)  # Vector of slopes
    c_slope = 0  # Scalar slope for intercept
    M = len(points)
    for i in range(M):
        x = points[i, :-1]  # All features except target
        y = points[i, -1]   # Target value
        error = y - (np.dot(m, x) + c)
        
        m_slope += (-2 / M) * error * x  # Element-wise multiplication
        c_slope += (-2 / M) * error
    
    # Update m and c
    new_m = m - learning_rate * m_slope
    new_c = c - learning_rate * c_slope
    return new_m, new_c
    return new_m, new_c
```

```python
def cost(points, m, c):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i, 0]
        y = points[i, 1]
        total_cost += (1/M) * ((y - m*x - c)**2)
    return total_cost
```

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler

def run():
    # Load the data and convert to NumPy array
    data = pd.read_csv('cali.csv').to_numpy()
    
    # Feature scaling (Standardization)
    scaler = StandardScaler()
    X = data[:, :-1]
    Y = data[:, -1]
    X = scaler.fit_transform(X)  # Normalize features
    data = np.hstack((X, Y.reshape(-1, 1)))  # Combine scaled features and target
    
    learning_rate = 1e-5  # Very small learning rate
    num_iterations = 20  # Increased iterations
    
    # Gradient descent to find m and c
    m, c = gd(data, learning_rate, num_iterations)
    print("Final m:", m)
    print("Final c:", c)

def gd(points, learning_rate, num_iterations):
    num_features = points.shape[1] - 1  # Exclude target column
    m = np.zeros(num_features)  # Initialize m as vector
    c = 0  # Initialize c as scalar
    
    for i in range(num_iterations):
        m, c = step_gradient(points, learning_rate, m, c)
        print(i, "Cost:", cost(points, m, c))  # Print total cost
    return m, c

def step_gradient(points, learning_rate, m, c):
    num_features = points.shape[1] - 1
    m_slope = np.zeros(num_features)  # Slope for m
    c_slope = 0  # Slope for c
    
    M = len(points)
    for i in range(M):
        x = points[i, :-1]  # Feature values
        y = points[i, -1]   # Target value
        error = y - (np.dot(m, x) + c)
        
        m_slope += (-2 / M) * error * x
        c_slope += (-2 / M) * error
    
    # Update m and c
    new_m = m - learning_rate * m_slope
    new_c = c - learning_rate * c_slope
    return new_m, new_c

def cost(points, m, c):
    total_cost = 0
    M = len(points)
    for i in range(M):
        x = points[i, :-1]
        y = points[i, -1]
        total_cost += (y - (np.dot(m, x) + c)) ** 2
    return total_cost / M

run()
```
