```python
documents = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        documents.append((movie_reviews.words(fileid), category))
documents[0: 5]
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import wordnet
def get_simple_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
from nltk import pos_tag
w = "better"
# pos_tag(w)
#we are supposed to pass an array
pos_tag([w])
```

```python
from nltk.corpus import stopwords
stops = set(stopwords.words('english'))
import string
punctuations = list(string.punctuation)
stops.update(punctuations)
stops
```

```python
def clean_review(words):
    output_words=[]
    for w in words:
        if w.lower() not in stops:
            pos = pos_tag([w])
            clean_words = lemmatizer.lemmatize(w, pos = get_simple_pos(pos[0][1]))
            output_words.append(clean_words.lower())
    return output_words
```

```python
documents = [(clean_review(document), category) for document, category in documents]
```

```python
documents[0]
```

```python
training_documents = documents[0:1500]
testing_documents = documents[1500:0]
```

```python
a = [1, 2]
b = [3, 4]
a += b
a
```

```python
all_words = []
for doc in training_documents:
    all_words += doc[0]
```

```python
import nltk
```

```python
freq = nltk.FreqDist(all_words)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
features
```

```python
def get_feature_dict(words):
    current_features = {}
    words_set = set(words)
    for w in features:
        current_features[w] = w in words_set
    return current_features
```

```python
get_feature_dict(training_documents[0][0])
```

```python
training_data = [(get_feature_dict(doc), category) for doc, category in training_documents]
testing_data = [(get_feature_dict(doc), category) for doc, category in training_documents]
```

```python
training_data[0]
```

```python
from nltk import NaiveBayesClassifier
```

```python
classifier = NaiveBayesClassifier.train(training_data)
```

```python
nltk.classify.accuracy(classifier, testing_data)
```
